2)How does broadcasting work in NumPy?
- Broadcasting in NumPy allows operations on arrays of different shapes without explicit replication, making computations more efficient. When performing operations between arrays of different dimensions, NumPy automatically expands the smaller array’s shape to match the larger one, following specific broadcasting rules.
- First, dimensions are aligned from the right. If shapes don’t match, dimensions with size 1 are stretched or duplicated. If incompatible shapes exist, an error occurs.
- Broadcasting optimizes memory usage and speeds up computations by avoiding redundant data copies.
- It’s widely used in element-wise operations, enabling efficient mathematical computations on arrays of varying shapes and sizes.

6)What are the differences between NumPy arrays and Python lists?
- NumPy arrays and Python lists differ in performance, functionality, and memory efficiency.
- NumPy arrays are fixed-type, homogeneous data structures optimized for numerical computations, whereas Python lists are dynamic and can store heterogeneous data. NumPy arrays provide faster operations due to vectorization and efficient memory usage, while lists require looping, making them slower for large-scale calculations. Arrays support element-wise operations, broadcasting, and advanced mathematical functions, which lists lack.
- Additionally, NumPy arrays consume less memory as they store elements in contiguous memory blocks, unlike lists, which store references. These advantages make NumPy arrays ideal for scientific computing and data analysis.

8)What does the term “vectorized operation” mean in NumPy?
- A vectorized operation in NumPy refers to performing element-wise computations on entire arrays without explicit loops. Unlike traditional Python loops, which process elements sequentially, vectorized operations leverage optimized C and Fortran libraries to execute computations in parallel, significantly improving speed and efficiency.
- NumPy achieves this by applying arithmetic, logical, and mathematical functions directly to arrays, reducing execution time and memory overhead. For example, adding two NumPy arrays (A + B) is faster than using a Python loop.
- Vectorization is essential for large-scale numerical computing, machine learning, and data analysis, making NumPy a powerful tool for high-performance applications.

4)How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
- Import necessary libraries
- Create or load a DataFrame
- Compute the correlation matrix
- Use Seaborn to visualize the heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [2, 3, 4, 5, 6],
    'C': [5, 3, 2, 8, 7],
    'D': [10, 12, 14, 16, 18]
}

# Create DataFrame
df = pd.DataFrame(data)

# Compute correlation matrix
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Show the plot
plt.title("Correlation Matrix Heatmap")
plt.show()
```

7)Write a program to perform element-wise multiplication of two NumPy arrays.

```python
import numpy as np

# Create two NumPy arrays
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([10, 20, 30, 40, 50])

# Element-wise multiplication using '*'
result1 = arr1 * arr2

# Element-wise multiplication using np.multiply()
result2 = np.multiply(arr1, arr2)

# Print results
print("Array 1:", arr1)
print("Array 2:", arr2)
print("Result using '*':", result1)
print("Result using np.multiply():", result2)
```

8)Create a line plot with multiple lines using Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values (common for all lines)
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10

# Generate y values for multiple lines
y1 = np.sin(x)  # Sine wave
y2 = np.cos(x)  # Cosine wave
y3 = np.tan(x)  # Tangent wave (clipped for visibility)

# Create the plot
plt.figure(figsize=(8, 5))  # Set figure size

plt.plot(x, y1, label="sin(x)", color="blue", linestyle="-", linewidth=2)
plt.plot(x, y2, label="cos(x)", color="red", linestyle="--", linewidth=2)
plt.plot(x, np.clip(y3, -2, 2), label="tan(x) (clipped)", color="green", linestyle="-.", linewidth=2)

# Customize the plot
plt.title("Multiple Line Plot using Matplotlib")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # Show legend
plt.grid(True)  # Add grid

# Show the plot
plt.show()
```

11) Perform matrix multiplication using NumPy

```python
import numpy as np

# Define two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication
result1 = np.dot(A, B)  # Using np.dot()
result2 = A @ B         # Using '@' operator

# Print results
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Result using np.dot():\n", result1)
print("Result using '@' operator:\n", result2)
```

13)Create a 3D scatter plot using Plotly.

```python
import plotly.express as px
import numpy as np
import pandas as pd

# Generate sample data
np.random.seed(42)  # For reproducibility
data = {
    'X': np.random.rand(50) * 10,  # Random values between 0 and 10
    'Y': np.random.rand(50) * 10,
    'Z': np.random.rand(50) * 10
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = px.scatter_3d(df, x='X', y='Y', z='Z', color='Z',
                     title="3D Scatter Plot using Plotly",
                     labels={'X': 'X Axis', 'Y': 'Y Axis', 'Z': 'Z Axis'})

# Show the plot
fig.show()
```
